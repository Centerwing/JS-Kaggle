```python
import pandas as pd

# Load a sample from each file
file_path = f"./datas/train.parquet/partition_id=4/part-0.parquet"
chunk = pd.read_parquet(file_path)

chunk = chunk[chunk["symbol_id"] == 1]

print(chunk.columns)
print(len(chunk))
```

```python
print(chunk.info())    
print(chunk.describe())  
print(chunk.head())   
print(chunk.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(24, 20))
corr = chunk.corr()
sns.heatmap(corr, cmap='coolwarm')#.get_figure().savefig("corr.png", dpi=500, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(24, 15))
sns.heatmap(chunk.isna())
plt.title("Missing Values Heatmap")
plt.show()
```
